```python
import cv2
```

```python
img = cv2.imread("image.jpg")

cv2.imshow("Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
img = cv2.imread("image.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.axis("off")
plt.show()
```

```python
print(img.shape)
print(gray.shape)
```

```python
resized = cv2.resize(img, (200, 30))

plt.imshow(resized)
plt.axis("off")
plt.show()
```
