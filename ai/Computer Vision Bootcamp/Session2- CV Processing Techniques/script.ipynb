```python
import cv2
import matplotlib.pyplot as plt
```

### Load the image

```python
img = cv2.imread("image.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

### thresholding

```python
ret, thresh = cv2.threshold(gray,44,255,cv2.THRESH_BINARY)

plt.imshow(thresh, cmap="gray")
plt.title("Binary Image")
plt.axis("off")
plt.show()
```

### Sobel 
| dx | dy | Result           |
| -- | -- | ---------------- |
| 1  | 0  | vertical edges   |
| 0  | 1  | horizontal edges |
| 1  | 1  | diagonal edges   |


```python
sobelx = cv2.Sobel(gray, cv2.CV_64F, 0,1)

plt.imshow(sobelx, cmap="gray")
plt.title("Sobel X")
plt.axis("off")
plt.show()
```

the Gx gave the best result for sobel method 

### Canny

```python
edges = cv2.Canny(gray,20,60)

plt.imshow(edges, cmap="gray")
plt.title("Canny Edges")
plt.axis("off")
plt.show()
```

i tried (300,200) (100,50) , (80,70),(20,30),(90,60),(20,60) and this one gave the best example

THE CANNY RESULT WAS BETTER THAN THE SOBEL METHOD

How do different Canny thresholds change the result?

the maxthreshold chooses the  gradiant value that we let it as edge(must be superior that the max)
while the minthrehold is to ignore the gradiant values less than it

```python
#experimenting the gaussian bluring
blur = cv2.GaussianBlur(gray,(5,5),0)

#for sobel first
sobelx = cv2.Sobel(blur, cv2.CV_64F, 0,1)

plt.imshow(sobelx, cmap="gray")
plt.title("Sobel X")
plt.axis("off")
plt.show()
```

```python
# for canny 

edges = cv2.Canny(blur,20,45)

plt.imshow(edges, cmap="gray")
plt.title("Canny Edges")
plt.axis("off")
plt.show()
```
