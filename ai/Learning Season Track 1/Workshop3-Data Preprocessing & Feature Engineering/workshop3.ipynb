# Handling missing values

```python
data = pd.DataFrame({ 'Age' : [14 , 25 , 20 , 19 , None , 26 , None]})
data
```

```python
# Mean
data["Age_mean"] = data["Age"].fillna(data["Age"].mean())

# Median
data["Age_median"] = data["Age"].fillna(data["Age"].median())
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
values = [[20], [40], [60], [80]]

normalized = scaler.fit_transform(values)
print(normalized)
```

```python
from sklearn.preprocessing import StandardScaler

values = [[20], [40], [60]]
scaler = StandardScaler()
standardized = scaler.fit_transform(values)
print(standardized)
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder

data = ["Low", "Medium", "High"]

encoder = LabelEncoder()
encoded = encoder.fit_transform(data)
print(encoded)
```

```python
# one-hot using pandas
data = pd.DataFrame({"Color": ["Red", "Blue", "Green", "Red"]})

one_hot = pd.get_dummies(data, columns=["Color"])
print(one_hot)
```

```python
# using sklearn
from sklearn.preprocessing import OneHotEncoder

data = pd.DataFrame({"Color": ["Red", "Blue", "Green"]})
encoder = OneHotEncoder()
encoded = encoder.fit_transform(data[['Color']])

print(data)
```

# Feature engineering

```python
df = pd.DataFrame({
    "Height": [1.7, 1.8],
    "Weight": [70, 80]
})

df["BMI"] = df["Weight"] / (df["Height"] ** 2)
print(df)
```

```python
df = pd.DataFrame({
    "User_ID": [101, 102],
    "Age": [20, 22]
})

df = df.drop(columns=["User_ID"])
print(df)
```
