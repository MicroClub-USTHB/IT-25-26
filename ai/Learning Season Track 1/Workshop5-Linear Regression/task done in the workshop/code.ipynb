```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , mean_absolute_error , r2_score
```

```python
df = pd.read_csv("Salary_Data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
plt.figure()
plt.scatter(df["YearsExperience"],df["Salary"])
plt.title("Salary vs year of experience ")
plt.xlabel("Year of experience ")
plt.ylabel("Salary")
plt.show()
```

```python
x = df[["YearsExperience"]]
y = df["Salary"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y, test_size=0.2, random_state = 42)
```

```python
salaryModel = LinearRegression()
```

```python
salaryModel.fit(x_train,y_train)
```

```python
predictions = salaryModel.predict(x_test)
```

```python
results = pd.DataFrame({
    "YearsExperience":x_test.values.flatten(),
    "rightsalaries":y_test.values,
    "Predicted_Salary":predictions
})
```

```python
results.to_csv("salary_prediction.csv", index = False)
```

```python
year = np.array([[5]])
prediction = salaryModel.predict(year)
```

```python
print("predicted salary of 5 years of ex : ",prediction[0])
```

```python
mae = mean_absolute_error(y_test,predictions )
mse = mean_squared_error(y_test,predictions)
r2 = r2_score(y_test,predictions)

print("mae : ",mae)
print("mse :", mse)
print("r2 : ",r2)
```

```python
plt.figure()
plt.scatter(x_train,y_train)
plt.plot(x_train,salaryModel.predict(x_train))
plt.title("linear regression line training data")
plt.xlabel("year of experience ")
plt.ylabel("salary")
plt.show()
```
